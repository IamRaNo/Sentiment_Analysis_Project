```python
slang_dict=dict(line.split("=", 1) for line in slangs.lower().strip().split("\n"))
```

```python
nlp.Defaults.stop_words
```

```python
words_to_retain = {
    "not", "no", "never", "none", "nobody", "nothing", "neither", "nor", 
    "nowhere", "but", "however", "although", "though", "yet", "unless", "without"
}
```

```python
stop_words = nlp.Defaults.stop_words - words_to_retain
```

```python
punctuations='"#$%&\'()*+,-./:;<=>@[\\]^_`{|}~'
```

```python
def pre_processing(text):
    text=text.lower()
    url_regex = re.compile(r"http[s]?://\S+|www\.\S+")
    text=url_regex.sub('', text)
    text=text.replace('`',"'").strip()
    text=text.strip()
    text=contractions.fix(text).lower()
    for item in text.split():
        if item in slang_dict.keys():
           text=text.replace(item,slang_dict.get(item))
    doc=nlp(text)
    processed_text=[]
    for tokens in doc:
        if tokens.text not in stop_words and tokens.text not in punctuations :
            processed_text.append(tokens.lemma_.lower())
    final_processed_text= [text for text in processed_text if text.strip()]
    return " ".join(processed_text)
```

```python
pre_processing('Hi nice to meet you')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf=TfidfVectorizer(max_features=10000,ngram_range=(1,2))
```

```python
vectors=tfidf.fit_transform(data['text'])
```

```python
vectors=vectors.toarray()
```

```python
vectors.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression,SGDClassifier
from sklearn.naive_bayes import MultinomialNB
from sklearn.ensemble import VotingClassifier,StackingClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import confusion_matrix,accuracy_score
```

```python
encoder=LabelEncoder()
```

```python
data['sentiment']=encoder.fit_transform(data['sentiment'])
```

```python
X_train,X_test,y_train,y_test=train_test_split(vectors,data['sentiment'],test_size=0.2)
```

```python
X_train.shape
```

```python
base_models = [
    ('lr',LogisticRegression(solver="saga", penalty="l2", C=5, max_iter=10000, tol=1e-4, n_jobs=-1)),
    ('mnb', MultinomialNB()),
    ('sgd',SGDClassifier(loss="log_loss", penalty="l2", alpha=1e-4, max_iter=5000, random_state=42))
]
```

```python
stacking_classifier = StackingClassifier(
    estimators=base_models,
    final_estimator=LogisticRegression()
)
```

```python
import time
start=time.time()
```

```python
stacking_classifier.fit(X_train,y_train)
```

```python
end=time.time()
print((end-start)/60)
```

```python
prediction=stacking_classifier.predict(X_test)
```

```python
print("Accuracy: ",accuracy_score(prediction,y_test))
```

```python
def predict_new(new_text):
    processed_text=pre_processing(new_text)
    processed_text=tfidf.transform([processed_text]).toarray()
    pred=stacking_classifier.predict(processed_text)
    return encoder.inverse_transform(pred)[0]
```
